```python
import gurobipy as gp
from gurobipy import GRB


model = gp.Model("Production Scheduling")

# Define the decision variables
x1 = model.addVar(vtype=GRB.INTEGER, name="x1")
x2 = model.addVar(vtype=GRB.INTEGER, name="x2")
x3 = model.addVar(vtype=GRB.INTEGER, name="x3")
x4 = model.addVar(vtype=GRB.INTEGER, name="x4")
x5 = model.addVar(vtype=GRB.INTEGER, name="x5")

# Set the objective function
model.setObjective(5*x1 + 4*x2 + 6*x3 + 7*x4 + 8*x5, GRB.MAXIMIZE)

# Add the constraints
model.addConstr(2*x1 + 3*x2 + x3 + 4*x4 + 5*x5 <= 100, "Labor")
model.addConstr(3*x1 + 2*x2 + 4*x3 + x4 + 3*x5 <= 80, "Machine Time")
model.addConstr(4*x1 + x2 + 3*x3 + 2*x4 + 2*x5 <= 70, "Raw Materials")
model.addConstr(x1 <= 20, "P1 Production Limit")
model.addConstr(x2 + x3 >= 10, "P2 and P3 Combined")
model.addConstr(x4 >= 5, "P4 Minimum Production")

# Optimize the model
model.optimize()

# Print the solution
print("Optimal Solution:")
print(f"x1 (P1): {x1.X:.0f}")
print(f"x2 (P2): {x2.X:.0f}")
print(f"x3 (P3): {x3.X:.0f}")
print(f"x4 (P4): {x4.X:.0f}")
print(f"x5 (P5): {x5.X:.0f}")
print(f"Total Profit: {model.ObjVal:.2f}")
```
